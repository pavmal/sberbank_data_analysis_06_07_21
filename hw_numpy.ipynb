### Упражнение 1

На одном графике изобразите время затраченное на вычисление с использованием массивов numpy и стандартных списков в питоне.  
Для более точных результатов замеряйте время выполнения функции несколько раз и усредняйте. 
* Используйте функцию `z = 2*x**2 + 4*y`
* Перемножение матриц размера n на n

````python
def func(x, y):
    return 2*x**2 + 4*y
````

````python
ls_x_out = []
ls_y_out = []
ar_x_out = []
ar_y_out = []

plt.figure(figsize=(8,4))
for num in np.arange(0, 10, 1):
    N = random.randint(5, 50)
    ls_x = [x for x in range(N)]
    ls_y = [y for y in range(N)]
    ar_x = np.array(ls_x)
    ar_y = np.array(ls_y)
    
    
    ls_tmp = []
    for j in range(3):
        calc_st = datetime.datetime.now()
        calc_func = [func(ls_x[i], ls_y[i]) for i in range(len(ls_x))]
        calc_end = datetime.datetime.now()
        ls_tmp.append((calc_end - calc_st).microseconds)
    ls_x_out.append(num)
    ls_y_out.append(np.mean(ls_tmp))
    
    ls_tmp = []
    for j in range(3):
        calc_st = datetime.datetime.now()
        calc_func = ar_x * ar_y
        calc_end = datetime.datetime.now()
        ls_tmp.append((calc_end - calc_st).microseconds)
    ar_x_out.append(num)
    ar_y_out.append(np.mean(ls_tmp))   
    
plt.plot(ls_x_out, ls_y_out, linewidth=2, color='red', marker='o', label='цикл')
plt.plot(ar_x_out, ar_y_out, linewidth=2, color='blue', marker='o', label='матричное')
plt.title("Сравнение скорости умножения в цикле с матричным")
plt.xlabel("Количество итераций")
plt.ylabel("Время, мл./сек.")
plt.legend(loc='upper right')
plt.show()
````

````python
ls_x_out = []
ls_y_out = []
ar_x_out = []
ar_y_out = []

plt.figure(figsize=(8,4))
for num in np.arange(0, 10, 1):
    N = random.randint(5, 50)
    ls_x = [[x for x in range(N)] for y in range(N)]
    ls_y = [[y for y in range(N)] for x in range(N)]
    ar_ls = np.zeros((N, N))
    ar_x = np.array(ls_x)
    ar_y = np.array(ls_y)
    
    
    ls_tmp = []
    for j in range(3):
        calc_st = datetime.datetime.now()
        for n in range(len(ls_x)):
            for m in range(len(ls_y)):
                ar_ls[n][m] = ls_x[n][m] * ls_y[n][m]
        calc_end = datetime.datetime.now()
        ls_tmp.append((calc_end - calc_st).microseconds)
    ls_x_out.append(num)
    ls_y_out.append(np.mean(ls_tmp))
    
    ls_tmp = []
    for j in range(3):
        calc_st = datetime.datetime.now()
        calc_func = ar_x * ar_y
        calc_end = datetime.datetime.now()
        ls_tmp.append((calc_end - calc_st).microseconds)
    ar_x_out.append(num)
    ar_y_out.append(np.mean(ls_tmp))   
    
plt.plot(ls_x_out, ls_y_out, linewidth=2, color='red', marker='o', label='цикл')
plt.plot(ar_x_out, ar_y_out, linewidth=2, color='blue', marker='o', label='матричное')
plt.title("Сравнение скорости умножения в цикле с матричным")
plt.xlabel("Количество итераций")
plt.ylabel("Время, мл./сек.")
plt.legend(loc='upper right')
plt.show()
````

### Упражнение 5
* Напишите функцию, которая берет многомерный вещественный массив с пропущенными значениями (np.nan) и возвращает его копию с заполненными пропусками (заменять средним значением всех элементов массива). Если в массиве одни пропущенные значения, заполните их нулями.

````python
def fill_mass(a):
    b = a.copy()
    if np.nansum(b) == 0:
        b[np.isnan(b)] = 0
    else:
        b[np.isnan(b)] = np.nanmean(b)
    return b 
````

````python
a = np.array([[1,2,3,np.nan,4], [6,np.nan,3,5,9]])
a
````

````python
b = fill_mass(a)
b
````

````python
a = np.array([[np.nan,np.nan,np.nan], [np.nan,np.nan,np.nan], [np.nan,np.nan,np.nan]])
a
````

````python
b = fill_mass(a)
b
````

### Упражнение 6
* Напишите функцию которая нормализирует заданный numpy-массив, так чтобы его значения лежали в интервале $[0, 1]$.

````python
def mass_normal(a):
    return (a - a.min()) / (a.max() - a.min())
````

````python
a = np.array([[1,2,3,5,4], [6,1,3,5,9]])
a
````

````python
mass_normal(a)
````

### Упражнение 9

* Напишите функцию, которая возвращает  $n \times m$ целочисленную матрицу с заданным значением на границах матрицы и с заданным значением внутри

````python
def new_matix(a, n, m):
    a[a>0] = n
    a[:1] = m
    a[-1:] = m
    a = a.T
    a[:1] = m
    a[-1:] = m
    a = a.T
    return a
````

````python
a = np.random.random((6, 4))
a
````

````python
new_matix(a, 5, 7)
````
